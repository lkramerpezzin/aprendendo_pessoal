#1) Runge-Kutta 
Método de Ralston para resolver o sistema de Lorenz
$$\frac{dX}{dt}= 10(Y-X)$$
$$\frac{dY}{dt}= X(28-Z) -Y$$
$$\frac{dZ}{dt}= XY - \frac{8}{3}Z$$
Com o tempo variando até  60: t=[0, 60]

##a)

```python
# Para X(0)= 0, Y(0)= 1 e Z(0)= 0

# Definindo as funções
def dx(y, x):
  return 10*(y-x)
def dy(y, x, z):
  return x*(28-z)-y
def dz(y, x, z):
  return x*y -(8/3)*z

# Definindo as variáveis
x0= 0
y0= 1
z0= 0
h= 0.001
t= 0

# Guardando valores
x_listA= [x0]
y_listA= [y0]
z_listA= [z0]
tempo= [t]


# RK-2 método de Ralston
while t<60:
  # em X
  k1_x= dx(y0, x0)
  k2_x= dx(y0+ (2/3)*h, x0+ (2/3)*k1_x*h)

  xn= x0+ h*((1/4)*k1_x + (3/4)*k2_x)
  
  x0= xn
  # em Y:
  k1_y= dy(y0, x0, z0)
  k2_y= dy(y0+ (2/3)*h, x0+ (2/3)*k1_y*h, z0)

  yn= y0+ h*((1/4)*k1_y + (3/4)*k2_y)
  
  y0= yn
  # em Z:
  k1_z= dz(y0, x0, z0)
  k2_z= dz(y0+ (2/3)*h, x0+ (2/3)*k1_z*h, z0)

  zn= z0+ h*((1/4)*k1_z + (3/4)*k2_z)
  
  z0= zn

  #
  t+= h 

  x_listA.append(x0)
  y_listA.append(y0)
  z_listA.append(z0)
  tempo.append(t)

###########C)

#Graficando
import matplotlib.pyplot as plt
plt.plot(tempo, x_listA, 'indigo')
plt.title('X do Sistema de Lorenz')
plt.xlabel('Tempo')
plt.ylabel('dX/dt')
plt.show()
plt.plot(tempo, y_listA, 'r')
plt.title('Y do Sistema de Lorenz')
plt.xlabel('Tempo')
plt.ylabel('dY/dt')
plt.show()
plt.plot(tempo, z_listA, 'g')
plt.title('Z do Sistema de Lorenz')
plt.xlabel('Tempo')
plt.ylabel('dZ/dt')
plt.show()
```
